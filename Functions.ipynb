<font color='steelblue'>Functions</font> 

A <font color='indianred'>function</font> is a reusable set of operations

The primary benefits to using functions are: 

 - **Reusability:** A function can be used over and over again without having to write redundant code. 
 
 - **Simplicity:** Functions are easy to use and make code readable. You only need to know the inputs and the purpose of the function. 
 
An input isn't even necessary, a function could perform its own computations to complete a task: 



<font color='steelblue'>Types of Functions</font> 

Two basic types of functions in Python: 

 - 1. Built - in functions
 - 2. User - defined functions
