 <font color='steelblue'>Conditional Statements</font>

A conditional statement is a Boolean expression that, if <font color='firebrick'>True</font>, executes a piece of code. 

Conditional statements allow programs to branch out into different paths based on Boolean expressions result in <font color='firebrick'>True</font> or <font color='firebrick'>True</font> 



Conditional statements control the *flow* of the code and allow the computer to *think* and so, they are called **control structures** 

Conditional statements are an integral part of programming 

```
if condition is True:
    #execute some code 
else:
    #execute some other code 
    pass 
```

Three types of conditional statements:

 - if
 - if - else
 - if - elif - else  

 <font color='steelblue'>if statements</font>
 
The *if* statement is comprised of two parts : 

 - the **condition**
 - the **code to be executed**
 
 
 
 While the <font color='firebrick'>:</font> is necessary to specifiy the beginning of the <font color='firebrick'>if</font> statement, the <font color='firebrick'>()</font> are optional. 
 
 An <font color='firebrick'>if</font> statement looks like this:
 
>**if** the **condition** holds <font color='firebrick'>True</font>,execute the **code to be executed**. 
Otherwise, **skip it and move on.


```
num = 5
if (num == 5): #the condition is true
    print('The number is equal to 5') #the code is executed
    
if num > 5: #the condition is false
    print('The number is greater than 5') #the code is not executed
```

 <font color='steelblue'>Conditions with Logical Operators</font>
 
 Logical Operators can be used to create more complex conditions in the  <font color='firebrick'>if</font> statement: 
 
 ```
 num = 12 
 if num % 2 == 0 and num % 3 == 0 and num % 4 == 0:
     #only works when num is a multiple of 2,3,and 4
     print('The number is a multiple of 2, 3, and 4')
    
if (num % 5 == 0 or num % 6 == 0):
    #only works when num is either a multiple of 5 or 6 
    print('The number is a multiple of 5 and/or 6')
```

 <font color='steelblue'>Nested if Statements</font>
 
 There can be if statements inside if statements
 
 ```
 num = 63
 
 if num >= 0 and num >= 100:
     if num >= 50 and num <= 75:
         if num >= 60 and num <= 70:
             print('The number is in the 60 - 70 range') 
``` 

 <font color='steelblue'>Creating and Editing Values</font>
 
 in a conditional statement you can edit the values of your variables as well as create new ones: 
 
 ```
 num = 10 
 
 if num > 5: 
     num = 20  #assigning a new value to num 
     new_num = num * 5 # creating a new value called new_num 
 
#the if condition ends, but the changes made inside it remain:
print(num)
print(new_num) 
```

 <font color='steelblue'>The if - else Statement</font>
 
  <font color='violet'>Structure</font>
  
  The  <font color='firebrick'>if - else</font> statement looks like :
  


If the **condition** turns out to be  <font color='indianred'>False</font>, the code after the <font color='indianred'>else:</font> keyword is executed

You can now perform two different actions based on the condition's value

The <font color='indianred'>else</font> keyword will be on the same indent level as the <font color='indianred'>if</font> keyword. It's body will be indented one tab to the right:

```
num = 60

if num <= 50:
    print('The number is less than or equal to 50')
else:
    print('The number is greater than 50')
```

<font color='steelblue'>Conditional Expressions</font>

*Conditional Expressions* use the functionality of an <font color='indianred'>if - else</font>statement in a different way - the expression returns an output based on the condition you provide which can be stored in a variable: 

```
output_value1 if condition else output_value2
```
If the <font color='indianred'>if</font>condition is fulfulled, the output would be <font color='indianred'>output_value1</font> otherwise, it would be <font color='indianred'>output_value2</font>

refactoring the above <font color='indianred'>if - else</font>statement:

```
num = 60

output = 'The number is less than or equal to 50'\
    if num <= 50 else "The number is greater than 50'
print(output)
```



<font color='steelblue'>The if - elif- else Statement</font>

The <font color='indianred'>if - else</font>statement handles two sides of the same condition: <font color='indianred'>True</font> and <font color='indianred'>False</font> which works very well if you are working with a problem that only has two possible outcomes

In programming, it isn't always a <font color='indianred'>True</font> or <font color='indianred'>False</font> situation, and a problem may have multiple outcomes which is where the <font color='indianred'>if - elif - else</font>statement comes in. It is the most comprehensive conditional statement because it allows you to create multiple conditions easily. 

The <font color='indianred'>elif</font> stands for **else - if** -- if the previous condition fails, thy this one instead. 

<font color='violet'>Structure</font>



```
light = 'red'

if light == 'green':
    print('go')
elif light == 'yellow':
    print('caution')
else:
    print('incorrect light signal')
```

<font color='steelblue'>Multiple elif Statements</font>

You can have as much <font color='indianred'>elif</font>s as you like, as long as they come between <font color='indianred'>if</font> and <font color='indianred'>else</font>



**NOTE** an <font color='indianred'>if - elif</font>statement can exist on its own without an <font color='indianred'>else</font> block at the end however, an <font color='indianred'>elif</font> cannot exist without an <font color='indianred'>if</font> statement preceeding it

```
num = 5 

if num == 0:
    print('zero')
elif num == 1:
    print('one')
elif num == 2:
    print('two)
elif num == 3:
    print('three')
elif num == 4:
    print('four')
elif num == 5:
    print('five')
elif num == 6:
    print('six')
elif num == 7:
    print('seven')
elif num == 8:
    print('eight')
elif num == 9:
    print('nine')
```
    
    
**NOTE** 
<font color='indianred'>if - elif - else </font> or <font color='indianred'>if - elif</font> statements are not the same as multiple <font color='indianred'>if</font> statements - <font color='indianred'>if</font>  statements act **independently** 

If the conditions of two successive <font color='indianred'>if</font>s are <font color='indianred'>True</font>, both statements will be executed

However, in <font color='indianred'>if - elif - else</font>, when a condition evaluates to <font color='indianred'>True</font>, the rest of the statement's conditions are not evaluated 

For example:

```
num = 10

if num > 5:
    print('The number is greater than 5')
    
if num % 2 == 0: 
    print('The number is even')

if not num % 2 == 0:
    print('The number is odd')
```
